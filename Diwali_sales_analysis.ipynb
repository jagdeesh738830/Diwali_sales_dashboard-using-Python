```python
kumar=sns.countplot(data = jagdeesh,x='Age Group',hue='Gender')
for bars in kumar.containers:
    kumar.bar_label(bars)
print('from Below graph we can see count of most the buyers between age gruop')
```

```python
sales_age= jagdeesh.groupby(['Age Group'], as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x= 'Age Group', y= 'Amount',data = sales_age)
```

```python
sns.countplot(data = jagdeesh,x= 'Age Group',hue='Gender')
```

```python
sales_state= jagdeesh.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state,x='State',y = 'Orders')
print('below graph represent that most of the state higher to buy the product')
```

```python
sales_state= jagdeesh.groupby(['State'], as_index = False)['Amount'].count().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
p=sns.barplot(data = sales_state,x='State',y = 'Amount')
for q in p.containers:
    p.bar_label(q)
```

```python
sales_state= jagdeesh.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].count().sort_values(by = 'Amount',ascending = False)
sns.set(rc={'figure.figsize':(10,5)})
a=sns.barplot(data = sales_state,x='Marital_Status',y = 'Amount',hue = 'Gender')
for b in a.containers:
    a.bar_label(b)
print('below graph represent that who is buy most product (married or onmarried)')
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
az = sns.countplot(data = jagdeesh,x = 'Occupation')
for bars in az.containers:
    az.bar_label(bars)

print('this graph is shows that buyers occupation')
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
az = sns.countplot(data = jagdeesh,x = 'Product_Category')
for bars in az.containers:
    az.bar_label(bars)

print('this graph represent that which product sales most')
```

```python
sales_state= jagdeesh.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc={'figure.figsize':(20,5)})
a=sns.barplot(data = sales_state,x='Product_Category',y = 'Amount')
for b in a.containers:
    a.bar_label(b)

print('below graph shows on that value of sum all product category wise')
```

```python

sales_state= jagdeesh.groupby(['Product_ID'], as_index = False)['Orders'].count().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x='Product_ID',y = 'Orders')

print('this graphs represent that what is the product id which is sales most')

    
```
